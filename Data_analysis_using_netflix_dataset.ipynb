#Question 4. show all the movies that were released in year 2020

```python
#we will create a new column called 'year' from the date_n column
#this will help use to filter movies or tv sereis using year
#we can also do the same for month as well
data['Year']=data['Date_N'].dt.year
data.head()
```

```python
data.isna()
```

```python
#since we are getting the year values as float type and our data set contains nan values we are unable to convert year column in to int type
#now we have to clean our data set
data=data.dropna(how='any')
data.isna().any()
```

```python
#now we have succesfully cleaned our data set
data['Year']=data['Year'].astype(int)
data.head()
```

```python
#NOW WE WILL FILTER THE MOVIES THAT WERE RELEASED IN YEAR 2020 FROM THE DATA FRAME
data[(data['Category']=='Movie')& (data['Year']==2020)]
```
